# Data Cleaning and Preprations

```python
# prepare for xgBoost handling categorical data

# List categorical columns
cat_cols = X_train.select_dtypes(include=['object']).columns
print(cat_cols)
# Convert both sets to 'category' dtype
for col in cat_cols:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')

```

# Machine Learning Model Training based on data analysis

```python
# XGBoost ML Model to train our regression model
from xgboost import XGBRegressor

model = XGBRegressor(
    tree_method="hist",        # Required for categorical support
    enable_categorical=True,   # Tells XGBoost to handle categories natively
    n_estimators=100,
    reg_lambda=1.0,  # Default L2; increase if still overfitting
    reg_alpha=0.1,
    objective='reg:squarederror',
    learning_rate=0.1, 
    max_depth=3
)

model.fit(X_train, y_train)
```

# Model Predictions

```python
y_pred = model.predict(X_test)

y_pred
```

# Save Results

```python
# 2. Create the new DataFrame
submission_df = pd.DataFrame({
    'Id': X_test['Id'],
    'SalePrice': y_pred
})

# Optional: View the first few rows
print(submission_df.head())
```
